```python
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
savings = [0,100,150,200,300,500,300,350,500,600,750,1000]
months = ['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec']
```

```python
plt.plot(months, savings) #plt.plt(x,y)

plt.xlabel('Month')

plt.ylabel('Savings ($)')

plt.title('Savings Account Results')

plt.show() #displays graph
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
#Change the color of the markers to red.
#Add a legend for the city of Chicago.
#Add a title and axes labels.
#Switch the axis so that the Fare($) data is on the x-axis.
#Add a limit for the x-axis data.
#Invert the y-axis so that January is at the top.


plt.scatter(y_axis, x_axis, color = 'r', label = 'Chicago')
plt.legend()
plt.xlabel('Fare ($)')
plt.ylabel('Month')
plt.title('PyBer Fare by Month')
plt.xlim(5,50)
plt.gca().invert_yaxis()
```

```python
plt.scatter(x_axis,y_axis, s = y_axis)
```

```python
#create for loop to make y_axis data points 3x larger
y_axis_larger = []
for data in y_axis:
    y_axis_larger.append(data*3)
    
plt.scatter(x_axis, y_axis, s = y_axis_larger)
```

```python
#create scatter plot with y_axis 3x larger via list comprehension
#new_list = [expression for item in list if conditional]

plt.scatter(x_axis,y_axis, s = [data * 3 for data in y_axis])
```

```python
#Create bubble chart OOP format
#Change the color of the markers to sky blue.
#Change the size of the markers to 5 times each data point.
#Make the color 20% transparent. alpha = .8
#Add a black edge color to the circles. edgecolor = 'k'
#Make the linewidth of the circles 2 points. linewidth = 2
#Add a legend for the city of Boston.
#Add a title.
#Switch the axis so that the Fare($) data is on the x-axis.
#Add a limit for the x-axis data.
#Invert the y-axis so that January is at the top.


#fig, ax = plt.subplots()
ax = plt.axes()

ax.scatter(y_axis, x_axis, color = 'skyblue', s = [y*5 for y in y_axis], alpha = .8, edgecolor = 'k', linewidth = 2, 
           label = 'Boston')
ax.legend()
ax.set_title('PyBer Fare by Month')
ax.set_xlim(5,50)
ax.set_xlabel('Fare ($)')
ax.set_ylabel('Month')
plt.gca().invert_yaxis()
```

```python
#Create pie chart
#plt.pie(y, labels, autopct, explode)
plt.subplots(figsize = (8,8))
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", 
          "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode_values = (0, 0, .1, 0, 0, 0, .2, 0, 0, 0, .3, 0)
plt.pie(y_axis, labels = x_axis, autopct = '%.1f%%', explode = explode_values, colors = colors)

plt.show()
```

```python
#Add a percentage to one decimal place to each wedge of the pie.
#Increase the figure size to 8x8. fig, ax = subplots(figsize = (8,8))
#Explode the two highest percentage months. explode = explode_values
#Add a shadow. shadow = True
#Add a start angle so that January is at the top. startangle = 90
#Reverse the order so that the month order is in a clockwise direction. counterclock = False
#Add new colors of your own choosing or use the colors from the previous pie chart.


colors = ["aqua", "tan", "coral", "crimson", "orangered", "lavender", 
          "lime", "silver", "gold", "sienna", "fuchsia", "ivory"]
explode_values = (0, 0, .1, 0, 0, 0, .2, 0, 0, 0, 0, 0)

fig, ax = plt.subplots(figsize = (8,8))
ax.pie(y_axis, labels = x_axis, autopct = '%.1f%%', explode = explode_values, colors = colors, shadow = True, startangle = 90,
      counterclock = False)

plt.show()
```
